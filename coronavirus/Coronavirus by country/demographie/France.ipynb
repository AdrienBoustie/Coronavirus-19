```python
import pandas as pd
```

```python
data=pd.read_csv("pop-totale-france-metro.csv")
data
```

```python
data=data.iloc[4:-1,1:]
```

```python
data=data.rename(columns={"Unnamed: 1":"Age","Unnamed: 4":"total"})
```

```python
data
```

```python
def clean(Series):
    for i in Series.index:
        Series=Series.replace(Series[i],"".join(Series[i].split()))   
    return Series   
```

```python
data1=pd.DataFrame(data.iloc[0:9,1]).agg(clean).astype('int64')      
data2=pd.DataFrame(data.iloc[10:19,1]).agg(clean).astype('int64')      
data3=pd.DataFrame(data.iloc[20:29,1]).agg(clean).astype('int64')      
data4=pd.DataFrame(data.iloc[30:39,1]).agg(clean).astype('int64')
data5=pd.DataFrame(data.iloc[40:49,1]).agg(clean).astype('int64')
data6=pd.DataFrame(data.iloc[50:59,1]).agg(clean).astype('int64')
data7=pd.DataFrame(data.iloc[60:69,1]).agg(clean).astype('int64')
data8=pd.DataFrame(data.iloc[70:79,1]).agg(clean).astype('int64')
data9=pd.DataFrame(data.iloc[80:,1]).agg(clean).astype('int64')

M=pd.DataFrame([data1.sum(),data2.sum(),data3.sum(),data4.sum(),data5.sum(),data6.sum(),data7.sum(),data8.sum(),data9.sum()],index=["0-9","10-19","20-29","30-39","40-49","50-59",'60-69',"70-79",'80+'])
```

```python
data.to_csv("france.csv")
```

```python
M=M.rename(columns={"Unnamed: 2":"population"})
M.loc[:,:]=M.loc[:,:]/M.iloc[:,0].sum()*100
```
