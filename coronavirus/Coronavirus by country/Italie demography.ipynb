```python
import numpy as np
import pandas as pd
```

```python
Age_italie=pd.read_csv("demographie/italy.csv")
Age_italie
```

```python
Age_italie=Age_italie.iloc[:,1:]
Age_italie[-1:]=0
```

```python
D=Age_italie.iloc[:,:].shift(1)
D=D.fillna(0)
for i in D.index :
    if i%2==0:
        D.loc[i,:]=np.nan
D
```

```python
for i in Age_italie.index :
    if i%2==0:
        Age_italie.loc[i,:]=0
Age_italie        
```

```python

M=D+Age_italie
M
```

```python
M=M.dropna()
M.index=["0-9","10-19","20-29","30-39","40-49","50-59","60-69","70-79","80+"]
M.loc[:,"Total"]=M.loc[:,"Nombre d'hommes"]+M.loc[:,"Nombre de femmes"]
M.loc[:,"rate"]=M.loc[:,"Total"]/M["Total"].sum()*100
M
```

```python
M.to_csv("demographie/italy_clean.csv")
```
